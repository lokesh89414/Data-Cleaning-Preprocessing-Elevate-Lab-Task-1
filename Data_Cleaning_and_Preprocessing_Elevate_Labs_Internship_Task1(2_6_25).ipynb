```python
# ----------------------------------------
# Step 4: Handling Missing Values
# ----------------------------------------
# Strategy: Fill missing categorical values with mode and numeric with mean
```

```python
for col in df.columns:
    if df[col].dtype == 'object':
        df[col].fillna(df[col].mode()[0], inplace=True)
    else:
        df[col].fillna(df[col].mean(), inplace=True)

print("\nMissing values after handling:")
print(df.isnull().sum())
```

```python
# ----------------------------------------
# Step 5: Removing Duplicates
# ----------------------------------------
```

```python
df.drop_duplicates(inplace=True)
```

```python
# ----------------------------------------
# Step 6: Standardizing Text Columns
# ----------------------------------------
# Example: Standardize 'Category' column if present
if 'Category' in df.columns:
    df['Category'] = df['Category'].str.strip().str.lower()
```

```python
# ----------------------------------------
# Step 7: Converting Date Columns
# ----------------------------------------
# Try converting any column that likely contains dates
for col in df.columns:
    if 'date' in col.lower():
        df[col] = pd.to_datetime(df[col], errors='coerce')
```

```python
# ----------------------------------------
# Step 8: Rename Columns for Consistency
# ----------------------------------------
df.columns = [col.strip().lower().replace(" ", "_") for col in df.columns]
```

```python
# ----------------------------------------
# Step 9: Fixing Data Types (if applicable)
# ----------------------------------------
# Example: Ensure 'Rating' is float and 'Reviews' is int
if 'rating' in df.columns:
    df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

if 'reviews' in df.columns:
    df['reviews'] = pd.to_numeric(df['reviews'], errors='coerce').fillna(0).astype(int)
```

```python
# ----------------------------------------
# Step 10: Export the Cleaned Dataset
# ----------------------------------------
df.to_csv('Google_data_cleaned.csv', index=False)
print("\n✅ Data cleaning complete. Cleaned file saved as 'Google_data_cleaned.csv'.")
```
